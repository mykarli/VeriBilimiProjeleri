```python
newData.Year = newData.Year.astype(str)
accidentNumber = newData.groupby("Year")
sum_accidentNumber=pd.DataFrame(accidentNumber.size().sort_values(ascending=False),columns=["SayÄ±"])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(y = sum_accidentNumber["SayÄ±"] , x = sum_accidentNumber.index)
plt.title("YÄ±llara gÃ¶re kaza sayÄ±sÄ±")
```

```python
data.Speed_limit = data.Speed_limit.astype(str)
speed_sayisi = data.groupby("Speed_limit")
toplam_speed=pd.DataFrame(speed_sayisi.size().sort_values(ascending=False),columns=["Speed_limit_toplam"])
sns.barplot(y = toplam_speed["Speed_limit_toplam"] , x = toplam_speed.index)
plt.title("HÄ±za gÃ¶re kaza sayÄ±sÄ±")
```
